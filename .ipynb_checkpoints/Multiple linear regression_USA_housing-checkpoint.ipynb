```python
sns.boxplot(y='Avg. Area Number of Bedrooms', data=df)
```

```python
df['Avg. Area Number of Bedrooms']=df['Avg. Area Number of Bedrooms'].fillna(df['Avg. Area Number of Bedrooms'].mean())
```

```python
df.isna().sum()
```

```python
import regex as re
```

```python
df['Address']
```

```python
def extract(data):
    return re.findall(r'\b[A-Z]{2}\b',data)[0]
```

```python
df['state'] = df['Address'].apply(extract)
df.head()
```

```python
df.drop(columns=['Address'],inplace=True)
df.head()
```

```python
def cap(data):
    q1=np.percentile(data,25)
    q3=np.percentile(data,75)
    iqr=q3-q1
    lt=q1-1.5*iqr
    ut=q3+1.5*iqr
    data.clip(lt,ut,inplace=True)
    
```

```python
outlier_col = df.columns
outlier_col=outlier_col.drop(['Avg. Area Number of Bedrooms','state'])
outlier_col
```

```python
for i in outlier_col:
    cap(df[i])
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(sc_x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
y_pred = model.predict(X_test)
```
